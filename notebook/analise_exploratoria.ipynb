```python
#filtrando colunas que vão ser usadas
cols_usar = [
    'data_assinatura_financiamento',
    'vlr_financiamento',
    'num_taxa_juros',
    'txt_tipo_imovel',
    'txt_sistema_amortizacao',
    'txt_regiao',
    'bln_cotista',
    'vlr_renda_familiar'
]

data_model = data_fgts_analitico[cols_usar].copy()
```

```python
#verificando informações das colunas
data_model.info()
data_model.shape
```

```python
#verificando a quantidade de valores nulos novamente
print(data_model.isnull().sum())
```

```python
#Tratando valores nulos

# Dropar linhas com data_assinatura_financiamento nula
data_model = data_model.dropna(subset=['data_assinatura_financiamento'])

# Criar flag para num_taxa_juros e preencher mediana
data_model['num_taxa_juros_na'] = data_model['num_taxa_juros'].isnull().astype(int)
data_model['num_taxa_juros'] = data_model['num_taxa_juros'].fillna(data_model['num_taxa_juros'].median())

# Criar flag para txt_sistema_amortizacao e preencher 'Desconhecido'
data_model['txt_sistema_amortizacao_na'] = data_model['txt_sistema_amortizacao'].isnull().astype(int)
data_model['txt_sistema_amortizacao'] = data_model['txt_sistema_amortizacao'].fillna('Desconhecido')

# Criar flag para bln_cotista e preencher 'Desconhecido'
data_model['bln_cotista_na'] = data_model['bln_cotista'].isnull().astype(int)
data_model['bln_cotista'] = data_model['bln_cotista'].fillna('Desconhecido')


data_model.head(10)
```

```python
print(data_model.isnull().sum())
```

```python
#Filtrando as linhas invalidas e outliers

# Remover financiamentos com valor 0 ou muito baixos (ex: <= 1000)
data_model = data_model[data_model['vlr_financiamento'] > 1000]

# Remover renda familiar zero ou nula
data_model = data_model[data_model['vlr_renda_familiar'] > 0]

data_model.head(10)
```

```python
#Melhorando informações de data

#Convertendo coluna para o tipo datetime do pandas
data_model['data_assinatura_financiamento'] = pd.to_datetime(data_model['data_assinatura_financiamento'], errors='coerce')

# Criando colunas ano e mês
data_model['ano'] = data_model['data_assinatura_financiamento'].dt.year
data_model['mes'] = data_model['data_assinatura_financiamento'].dt.month

# Removendo linhas com datas inválidas
data_model = data_model.dropna(subset=['data_assinatura_financiamento'])

data_model.head(10)
```

```python
label_cols = ['txt_tipo_imovel', 'txt_sistema_amortizacao', 'txt_regiao', 'bln_cotista']

for col in label_cols:
    le = LabelEncoder()
    data_model[col] = le.fit_transform(data_model[col])

data_model.head(10)
```

```python
print(data_model.head())
print(data_model.info())
print(data_model.describe())
```

```python
#Salvando dados tratados
data_model.to_csv('../data/processed/dados_fgts_tratados.csv', index=False)
```
